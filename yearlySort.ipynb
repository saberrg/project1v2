```python
import pandas as pd

#change dt.year and output file name 
file_path = 'Resources/15to23.csv'
output_file_path = 'Resources/2015.csv' 

chunk_size = 10000

# Create an empty DataFrame to store the rows with the specified year
year__df = pd.DataFrame()

# Iterate over the chunks of the CSV file (since it is a large datasets)
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    chunk['Date'] = pd.to_datetime(chunk['Date'])
    chunk_of_year = chunk[chunk['Date'].dt.year == 2015]
    
    # Append the rows with the year 2023 to the DataFrame
    year_df = year_df.append(chunk_of_year, ignore_index=True)

# Save the DataFrame with the rows of year (2015-2023) to a CSV file
year_df.to_csv(output_file_path, index=False)
```
